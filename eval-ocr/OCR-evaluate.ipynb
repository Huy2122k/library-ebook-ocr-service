```python
wers_arr = []
levenshteins_arr = []
cers_arr = []
for file, predict in tqdm(dict_res_ocr.items()):
    if not predict:
        continue
    file_text = rf'C:\Users\Vhc\Downloads\fiftyone\SROIE2019\test\box\{file}.txt'
    f = open(file_text, "r")
    text = ""
    for box in f.readlines():
        text+= box.split(",")[-1].lower()
    wers_arr.append(wer(text,predict[:len(text)]))
    levenshteins_arr.append(fuzz.ratio(text,predict[:len(text)]))
    cers_arr.append(cer(text,predict[:len(text)]))
```
